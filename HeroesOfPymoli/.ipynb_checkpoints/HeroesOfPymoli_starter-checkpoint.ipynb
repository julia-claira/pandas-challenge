```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#player count----------
totalPlayers=len(purchase_data["SN"].unique()) #finds the unique number of players
total_df=pd.DataFrame({"Total Players": [totalPlayers]}) #summarizes result in a data frame

#outputs result------------------
total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#analyzes purchasing data--------
numItems = len(purchase_data["Item ID"].unique()) #finds the unique items
totalPrice = purchase_data.loc[:,"Price"].sum() #finds the total price of all items
totalSales = purchase_data["Purchase ID"].count() #totals the number of sales
averagePrice = round(totalPrice/totalSales,2) #finds the average price per sale


#outputs the results in a data frame--------------
unique_items_table = pd.DataFrame({"Number of Unique Items": [numItems],
                         "Average Price": (f"${averagePrice}"),
                         "Number of Purchases":[totalSales],
                         "Total Revenue": (f"${totalPrice:,.2f}")})

#outputs result------------------
unique_items_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#analyzes basic Gender Data------------
uniquePlayer_df=purchase_data.drop_duplicates(subset=['SN']) #gets rid of rows with duplicate users
numGenders=uniquePlayer_df["Gender"].value_counts() #value count of all genders
gendPercent=round(numGenders/totalPlayers*100,2) #calculates the percentage of players for each gender

#creates data frame for output------------------
gender_df=pd.DataFrame({"Total Count":numGenders,
                       "Percentage of Players": gendPercent}) #creates a data frame with gender count
gender_df["Percentage of Players"]=gender_df["Percentage of Players"].astype('str')+"%" #formats percentage column

#outputs result------------------
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#analyzes purchasing data based on Gender---------------------
purchaseCount=purchase_data["Gender"].value_counts() #finds the num of purchases for each gender
genderGB=purchase_data.groupby("Gender") #groups the data by gender
genderPrice=round(genderGB.sum()["Price"],2) #finds the total spent for each gender
genderAverage=genderGB.mean()["Price"].round(decimals=2) #finds the average purchase price per gender
avgGenderPrice=round(genderPrice/numGenders,2) #the average purchase price per person

#formats results for output-----------------
genderAverage=genderAverage.map("${:,.2f}".format)
genderPrice=genderPrice.map("${:,.2f}".format)
avgGenderPrice=avgGenderPrice.map("${:,.2f}".format)

#setus up data frame table for output------------------------
purchaseAnaly_df=pd.DataFrame(index=genderAverage.index)
purchaseAnaly_df["Purchase Count"]=purchaseCount
purchaseAnaly_df["Average Purchase Price"]=genderAverage
purchaseAnaly_df["Total Purchase Price"]=genderPrice
purchaseAnaly_df["Avg Total Purchase per Person"]=avgGenderPrice

#outputs result------------------
purchaseAnaly_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#sorts unique players into bins based on age-------------
bins = [0,9,14,19,24,29,34,39,120] #bins range to sort age
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"] #bins category
uniquePlayer_df["Age Range"]=pd.cut(uniquePlayer_df["Age"],bins,labels=group_names,include_lowest=True)

#analayzes age range into new data frame-------------------
ageSummary_df=pd.DataFrame(uniquePlayer_df.groupby("Age Range").count()) #creates data frame with Age Bins as index
ageSummary_df=ageSummary_df.rename(columns={"Purchase ID":"Total Count"}) #changes column heading
ageSummary_df=ageSummary_df[["Total Count"]] #gets rid of unnecessary columns and only keeps 'Total Count'
ageSummary_df["Percentage of Players"]=ageSummary_df["Total Count"]/totalPlayers*100 #finds age range percentage and creates new column
ageSummary_df["Percentage of Players"]=ageSummary_df["Percentage of Players"].map("{:.2f}%".format) #formats percentage output

#outputs result------------------
ageSummary_df 
```

* Optional: give the displayed data cleaner formatting

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#groups data frame by name handle
user_group=purchase_data.groupby("SN")

#totals the purchases per person, total price per person, and average price per person
totalSpending=user_group.sum() #creates data frame and calculates total spending per person
totalSpending["Purchase Count"]=user_group["Purchase ID"].count() #adds column with purchase count
totalSpending["Average Purchase Price"]=totalSpending["Price"]/totalSpending["Purchase Count"] #adds column with average price per person

#sorts rows by most biggest spender and only keeps the top 5
totalSpending=totalSpending.sort_values("Price",ascending=False).head(5)

#creates a new data frame with the final data
topSpenders_df=totalSpending[["Purchase Count","Average Purchase Price","Price"]]

#renames the last column
topSpenders_df=topSpenders_df.rename(columns={"Price":"Total Purchase Value"})

#formats the columns
topSpenders_df["Average Purchase Price"]=topSpenders_df["Average Purchase Price"].map("${:,.2f}".format)
topSpenders_df["Total Purchase Value"]=topSpenders_df["Total Purchase Value"].map("${:,.2f}".format)

#outputs data frame
topSpenders_df
```
